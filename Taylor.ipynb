```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
# Sytem parapeters
ω   = 2*np.pi
ω_ο = 1.5*ω
β = ω_ο/4 
γ = 0.2

# Simulation parameters
N = 1000
t = np.linspace(0,6,N)
# Initial conditions
θ = [0,0]
def f(θ,t):
    dvdt = -ω_ο**2*np.sin(θ[0])-2*β*θ[1]+γ*ω_ο**2*np.cos(ω*t)
    dθdt = θ[1]
    return[dθdt, dvdt]
sol = odeint(f,θ,t)
θs = sol[:,0]

plt.plot(t,θs)
plt.show()
```

```python
# Biffurcation
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
# Sytem parapeters
ω   = 2*np.pi
ω_ο = 1.5*ω
β = ω_ο/4 

# Simulation parameters
dt = 1e-4
t = np.arange(0,500,dt)
# Initial conditions
θ = [-np.pi/2,0]

def f(θ,t):
    dvdt = -ω_ο**2*np.sin(θ[0])-2*β*θ[1]+γ*ω_ο**2*np.cos(ω*t)
    dθdt = θ[1]
    return[dθdt, dvdt]
    
γ_m = np.arange(1,2.5,0.0001)
bif = np.zeros([len(γ_m),100])
for j in range(len(γ_m)):
    γ = γ_m[j]
    sol = odeint(f,θ,t)
    bif[j,:] = sol[int(400/dt):int(500/dt):int(1/dt),1]
    
plt.plot(γ_m,bif,'k.',markersize=1)
plt.show()
```

```python
# Biffurcation
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
# Sytem parapeters
ω   = 2*np.pi
ω_ο = 1.5*ω
β = ω_ο/4 

# Simulation parameters
dt = 1e-3
t = np.arange(0,600,dt)
# Initial conditions
θ = [-np.pi/2,0]

def f(θ,t):
    dvdt = -ω_ο**2*np.sin(θ[0])-2*β*θ[1]+γ*ω_ο**2*np.cos(ω*t)
    dθdt = θ[1]
    return[dθdt, dvdt]
    
γ_m = np.arange(1.06,1.087,0.00001)
bif = np.zeros([len(γ_m),100])
for j in range(len(γ_m)):
    γ = γ_m[j]
    sol = odeint(f,θ,t)
    bif[j,:] = sol[int(500/dt):int(600/dt):int(1/dt),0]
    
plt.plot(γ_m,bif,'k.',markersize=1)
plt.show()
```
