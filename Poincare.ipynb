```python
# Biffurcation
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
# Sytem parapeters
T   = 1
ω   = 2*np.pi*T
ω_ο = 1.5*ω
β   = ω_ο/4
γ   = 1.078

# Simulation parameters
dt = 1e-2
N  = 600
n = int(1/dt)
t = np.arange(0,N*T,dt)
# Initial conditions
θ = [-np.pi/2,0]

def f(θ,t):
    dvdt = -ω_ο**2*np.sin(θ[0])-2*β*θ[1]+γ*ω_ο**2*np.cos(ω*t)
    dθdt = θ[1]
    return[dθdt, dvdt]
    
sol = odeint(f,θ,t)

X = sol[10*n:60*n:n,0]
Y = sol[10*n:60*n:n,1]

plt.plot(X,Y,'k.',markersize=1)
plt.show()
```
