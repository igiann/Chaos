```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
# Sytem parapeters
fn = 1.5
[ωN] = [2*np.pi*fn]
# Simulation parameters
N = 1000
t = np.linspace(0,6,N)
# Initial conditions
θ = [0.1,0]
def f(θ,t):
    dvdt = -ωN**2*np.sin(θ[0])
    dθdt = θ[1]
    return[dθdt, dvdt]
sol = odeint(f,θ,t)
θs = sol[:,0]
ωs = sol[:,1]

plt.plot(t,θs)
plt.show()
plt.plot(t,ωs)
plt.show()
plt.plot(θs/np.pi,ωs)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
# Sytem parapeters
fn = 1.5
[ωN] = [2*np.pi*fn]
β = 0.5
# Simulation parameters
N = 1000
t = np.linspace(0,6,N)
# Initial conditions
θ = [0.1,0]
def f(θ,t):
    dvdt = -ωN**2*np.sin(θ[0])-2*β*θ[1]
    dθdt = θ[1]
    return[dθdt, dvdt]
sol = odeint(f,θ,t)
θs = sol[:,0]
ωs = sol[:,1]

plt.subplot(3,1,1)
plt.plot(t,θs)
plt.subplot(3,1,2)
plt.plot(t,ωs)
plt.subplot(3,1,3)
plt.plot(θs/np.pi,ωs)
plt.show()
```
